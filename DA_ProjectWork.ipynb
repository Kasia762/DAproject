```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pandas.util.testing as tm
import statsmodels.api as sm
import seaborn as sns
import statsmodels.formula.api as smf
```

```python
df = pd.read_csv('OnlineNewsPopularity.csv')
listA = list(range(2,61))
print(listA)
df1 = df.iloc[:,listA]
df1.columns = df1.columns.str.replace("_","")
#df1.head()
```

```python
dftrain = df1.sample(frac=0.8, random_state = 200) #training data set (80%)
dftest = df1.drop(dftrain.index) #testing data set (20%)
```

```python
sns.set(rc = {'figure.figsize':(15,8)})
sns.set_theme(style="white")
corr = dftrain.corr()
#corr = corr.iloc[[1,2,3,4,32,33,34,35],[1,2,3,4,32,33,34,35]]
#dfCorr = dftrain.corr()
#fdf = dfCorr[((dfCorr >= .5) | (dfCorr <= -.5)) & (dfCorr !=1.000)]
#corr = fdf[fdf.notna()]

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Custom colormap
cmap = sns.diverging_palette(275, 150, s=90, l=50, n=9, as_cmap=True)

sns.heatmap(corr, mask=mask, cmap=cmap, vmax=1, linewidths=0.3, cbar_kws={"shrink":0.5})
```
