```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pandas.util.testing as tm
import statsmodels.api as sm
import seaborn as sns
import statsmodels.formula.api as smf
```

```python
df = pd.read_csv('s41598-020-73558-3_sepsis_survival_primary_cohort.csv')
df.head()
```

```python
sns.set(rc = {'figure.figsize':(15,8)})
sns.set_theme(style="white")
corr = df.corr()

mask = np.triu(np.ones_like(corr, dtype=bool))

cmap = sns.diverging_palette(275, 150, s=90, l=50, n=9, as_cmap=True)

sns.heatmap(corr, mask=mask, cmap=cmap, vmax=1, linewidths=0.3, cbar_kws={"shrink":0.5})
```
